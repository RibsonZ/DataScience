```python
X = df[df.columns[1:17]]  # zmienne niezalezne, bez 'Name' i solubility
Y = LogS #zmienna zalezna
```

```python
X
```

```python
# tworzymy model regresji liniowej f(x) = w0 + w1*x1 + ... + wn*xn, bedziemy manipulowac wagami tak,
# by uzyskac nabardziej adekwatna linie (najmniejsze kwadraty, LINEST in excel)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X, Y)
```

```python
y_pred = reg.predict(X)
```

```python
results = pd.DataFrame(y_pred, columns = ["Predicted"])
results["Observed"] = Y
results
```

```python
SSE = np.sum(np.power(results.Observed - results.Predicted, 2))
SSE
```

```python
MSE = np.sum(np.power(results.Observed - results.Predicted, 2))/results.shape[0]
MSE
```

```python
Y.values.max() - Y.values.min()
```

```python
RMSE = np.sqrt(MSE)
NRMSE = RMSE / (Y.values.max() - Y.values.min())*100
```

```python
RMSE
```

```python
NRMSE
```

```python
# wspolczynnik determinacji, jak duza czesc zmiennosci wyjasnia nasz model
print("Learn metrics")
print("R2", metrics.r2_score(Y, y_pred))
print("RMSE", metrics.mean_squared_error(y_pred, Y, squared = False))
print("NRMSE", metrics.mean_squared_error(y_pred, Y, squared = False)/(Y.values.max() - Y.values.min())*100)
```

```python
# regularyzacja - karanie modelu za skomplikowanie
# regresja Lasso
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(
                                    X, Y, test_size = 40, shuffle = True)


regLasso = linear_model.Lasso(alpha = 0.19)
regLasso.fit(X_train, y_train)
y_pred = regLasso.predict(X_test)

print("Learn metrics")
print("R2", metrics.r2_score(y_test, y_pred))
print("RMSE", metrics.mean_squared_error(y_pred, y_test, squared = False))
print("NRMSE", metrics.mean_squared_error(y_pred, y_test, squared = False)/(y_test.values.max() - y_test.values.min())*100)
```

```python
for i in np.arange(0.05, 0.5, 0.05):
    regLasso = linear_model.Lasso(alpha = i)
    regLasso.fit(X_train, y_train)
    y_pred = regLasso.predict(X_test)
    print("Learn metrics")
    print("R2", metrics.r2_score(y_test, y_pred))
    print("RMSE", metrics.mean_squared_error(y_pred, y_test, squared = False))
    print("NRMSE", metrics.mean_squared_error(y_pred, y_test, squared = False)/(y_test.values.max() - y_test.values.min())*100)
```
